```python
import seaborn as sns
import matplotlib.pyplot as plt

# Anzahl der gewünschten Farben
n_colors = 5

# Farben aus der Colormap abrufen
colors = sns.color_palette("Spectral", n_colors=n_colors)
colors = sns.color_palette("husl", 5)

# Beispielplot, um die Farben zu überprüfen
plt.scatter(range(n_colors), [1] * n_colors, c=colors, marker='o', s=200)
plt.show()
```
